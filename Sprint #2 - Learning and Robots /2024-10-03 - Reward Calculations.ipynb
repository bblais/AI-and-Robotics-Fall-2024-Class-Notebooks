```python
Q=0
```

```python
Qs=[Q]
for i in range(100):
    reward=+1
    Q=Q+α*(reward-Q)

    Qs.append(Q)

plot(Qs,'-o')
```

```python
Q=0
Qs=[Q]
for i in range(100):
    reward=-1
    Q=Q+α*(reward-Q)

    Qs.append(Q)

plot(Qs,'-o')
```

```python
Q=0
Qs=[Q]
for i in range(100):
    r=rand()  # between 0 and 1
    if r<0.1: # 10% of the time
        reward=-1
    else:
        reward=+1
    
    Q=Q+α*(reward-Q)

    Qs.append(Q)

plot(Qs,'-o')
```
